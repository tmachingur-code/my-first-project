# DATA TYPES

```python
#INTEGERS ~ Positive whole number, negative whole number and zero (eg -2, 2)
#FLOATS ~ Numbers with decimals (eg 2.5, 3.14)
#Boolean ~ represents the true values and can be only True or False
# Lists  ~ Lists are ordered, mutable collections of items with square brackets []
#Strings ~ used for text. they are immutable
#Tuples ~  ares ordered, immutable collections with parentheses().
#Range ~ is a sequence of numbers used mainly in loops. It does not store all numbers in memory.
#Dictionaries ~ Stores data in Key-value pairs using curly braclets {}.
```

```python
#Integers

x = 100
y = 20
print (x+y)
```

```python
#Strings
name = "Tsungi"
print (name)
print(type(name))
```

```python
#Boolean

is_valid = True
print (is_valid)
print(type(is_valid))
```

```python
#Lists

shopping = ["Eggs", "Milk", "Bread"]
print(shopping)
print(type(shopping))
```

```python
#Tuples
coordinates = (10, 20)
print(coordinates)
print(type(coordinates))
```

```python
#Range
nums = range(5)
print(list(nums))
print(type(nums))
```

```python
#Dictionaries
phonebook = {"Alice": "555-1234", "Bob": "555-9876"}
print(phonebook)
print(type(phonebook))
```
